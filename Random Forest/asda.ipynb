```python
import pandas as pd
import numpy as np

data = pd.read_csv('emails.csv')
data.head()
```

```python
data.info()
```

```python
data=data.drop(columns="Email No.")
data
```

```python
data['Prediction'].value_counts() # 1은 spam
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
X_train = train_data.iloc[:, 1:-1]
y_train = train_data.iloc[:, -1]


X_test = test_data.iloc[:, 1:-1]
y_test = test_data.iloc[:, -1]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# RandomForestClassifier 인스턴스 생성
model = RandomForestClassifier(n_estimators=500,oob_score=True,random_state=42, n_jobs=-1)
param_grid = {'max_depth': [5, 10, 15, 20]}

# Grid Search
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Print the best parameters from Grid Search
print("Best Parameters from Grid Search:", grid_search.best_params_)
model_best_grid = grid_search.best_estimator_

importance = model_best_grid.feature_importances_
print(importance)
```

```python
sorted_indices = np.argsort(importance)[::-1]
print(grid_search.best_score_)
# 상위 N개의 피처 출력
top_n_features = 5
selected_features = sorted_indices[:top_n_features]

print("Top", top_n_features, "Features:")
for idx in selected_features:
    print(f"Feature {idx} : {data.columns[idx]} Importance = {importance[idx]}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# 테스트 데이터셋에 대한 예측
y_pred = model_best_grid.predict(X_test)

# 정확도 계산
accuracy = accuracy_score(y_test, y_pred)

# 혼동 행렬 계산
confusion_mat = confusion_matrix(y_test, y_pred)
print(accuracy)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 혼동 행렬 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', xticklabels=['정상', '스팸'], yticklabels=['정상', '스팸'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

```
